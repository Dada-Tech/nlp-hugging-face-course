```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
```

```python
classifier(["This movie was ahead of its time", "i'd probably go again", "im only coming here if im extremely hungry"])
```

```python
classifier = pipeline("zero-shot-classification")
classifier(
    "Neo-Soul genre is really similar to RnB and more melodic than hip-hop, but much easier to play than other genres like classical or heavy metal",
    candidate_labels=["education", "politics", "business", "guitar"],
)
```

```python
generator = pipeline("text-generation", model="distilgpt2")
generator(
    "Hey boss, sorry I cant come into the office today, I'm",
    max_length=30,
    num_return_sequences=4,
)
```
